```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 40)
pd.set_option('display.max_columns', 40)
```

```python
van_df = pd.read_csv('Vancouver_Location_Final.csv')
den_df = pd.read_csv('Denver_Location_Final.csv')
```

```python
merged_location_df = pd.concat([van_df, den_df], ignore_index=True, sort=False) # Concatenates both tables
```

```python
# Removing quotation marks
merged_location_df.iloc[:, 1] = merged_location_df.iloc[:, 1].str.replace('"', '')
```

```python
merged_location_df.to_csv('Merged_Location_Final.csv', index=False)
```
