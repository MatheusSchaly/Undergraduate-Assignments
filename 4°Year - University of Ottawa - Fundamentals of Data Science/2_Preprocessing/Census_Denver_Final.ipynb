```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 60)
pd.set_option('display.max_rows', 4)
```

```python
df = pd.read_csv("Census_Denver.csv")
```

```python
# Drops useless columns
df.columns.tolist()
df = df.drop(df.columns[47:], axis=1)
df = df.drop(df.columns[3:24], axis=1)
df = df.drop(df.columns[0], axis=1)
```

```python
df.head()
```

```python
# Merge some columns
df['Years-15-to-19'] = df['AGE_15_TO_17'] + df['AGE_18_AND_19']
df['Years-20-to-24'] = df['AGE_20'] + df['AGE_21'] + df['AGE_22_TO_24']
df['Years-60-to-64'] = df['AGE_60_AND_61'] + df['AGE_62_TO_64']
df['Years-65-to-69'] = df['AGE_65_AND_66'] + df['AGE_67_TO_69']
df.drop(['AGE_15_TO_17', 'AGE_18_AND_19', 'AGE_20', 'AGE_21', 'AGE_22_TO_24',
         'AGE_60_AND_61', 'AGE_62_TO_64', 'AGE_65_AND_66', 'AGE_67_TO_69'], axis=1, inplace=True)
```

```python
# Rename columns
df = df.rename(columns={'NBRHD_NAME': 'Neighborhood', 'POPULATION_2010': 'Total-neighborhood-population',
                       'AGE_LESS_5': 'Years-0-to-4', 'AGE_5_TO_9': 'Years-5-to-9',
                       'AGE_10_TO_14': 'Years-10-to-14', 'AGE_25_TO_29': 'Years-25-to-29', 
                       'AGE_30_TO_34': 'Years-30-to-34', 'AGE_35_TO_39': 'Years-35-to-39', 
                       'AGE_40_TO_44': 'Years-40-to-44', 'AGE_45_TO_49': 'Years-45-to-49', 
                       'AGE_50_TO_54': 'Years-50-to-54', 'AGE_55_TO_59': 'Years-55-to-59',
                       'AGE_70_TO_74': 'Years-70-to-74', 'AGE_75_TO_79': 'Years-75-to-79', 
                       'AGE_80_TO_84': 'Years-80-to-84', 'AGE_85_PLUS': 'Years-85-plus'}) # Formats names
```

```python
# Repositioning Columns
cols = df.columns.tolist()
df = df[cols[0:5] + cols[16:18] + cols[5:12] + cols[18:20] + cols[12:15] + cols[15:16]]

df.set_index('Neighborhood', inplace=True) # Neighborhood is now the index
df = df.transpose() # Transpose the dataframe
```
